Read Data

```python
import pandas as pd
import numpy as np

data=pd.read_csv('insurance.csv')
data.head(n=6)
```

EDA

```python
import matplotlib.pyplot as plt

```

```python
plt.hist(data['charges'])
```

```python
plt.hist(np.log(data['charges']),bins=20)
```

Feature Engineering

```python
data=pd.get_dummies(data)
data.head()
```

```python
x=data.drop(['charges'], axis=1)
x
```

```python
y=data['charges']
y
```

```python
x.fillna(0,inplace=True)
y.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler(with_mean=True,with_std=True)
scaler=scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
x_train_scaled
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features=PolynomialFeatures(degree=2,include_bias=False)
x_train_transform=poly_features.fit_transform(x_train_scaled)
x_test_transform=poly_features.fit_transform(x_test_scaled)

x_train_transform
```

Train the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

reg=LinearRegression()

reg.fit(x_train_scaled,np.log1p(y_train))
y_predict=reg.predict(x_test_scaled)

# ucsd ridge:

ridge=Ridge()
ridge.fit(x_train_scaled,np.log1p(y_train))
y_predict_ridge=ridge.predict(x_test_scaled)
```

模型评估

```python
from sklearn.metrics import mean_squared_error

log_rmse_train=np.sqrt(mean_squared_error(y_true=np.log1p(y_train),y_pred=reg.predict(x_train_scaled)))
log_rmse_test=np.sqrt(mean_squared_error(y_true=np.log1p(y_test),y_pred=y_predict))
rmse_train=np.sqrt(mean_squared_error(y_true=y_train,y_pred=np.exp(reg.predict(x_train_scaled))))
rmse_test=np.sqrt(mean_squared_error(y_true=y_test,y_pred=np.exp(reg.predict(x_test_scaled))))

print(log_rmse_train,log_rmse_test,rmse_train,rmse_test)

#evaluate ridge

log_rmse_train1=np.sqrt(mean_squared_error(y_true=np.log1p(y_train),y_pred=ridge.predict(x_train_scaled)))
log_rmse_test1=np.sqrt(mean_squared_error(y_true=np.log1p(y_test),y_pred=y_predict_ridge))
rmse_train1=np.sqrt(mean_squared_error(y_true=y_train,y_pred=np.exp(ridge.predict(x_train_scaled))))
rmse_test1=np.sqrt(mean_squared_error(y_true=y_test,y_pred=np.exp(ridge.predict(x_test_scaled))))

print(log_rmse_train1,log_rmse_test1,rmse_train1,rmse_test1)
```
